```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [0,1,2,3,4,5,6]
y = [45,32,56,76,87,54,32]
```

```python
plt.plot(x, y)
```

```python
# change graph line color, marker and style
plt.plot(x, y, color = 'red', marker = 'o', linestyle = '--')
```

```python
# add text for each point in the graph
plt.plot(x, y, color = 'red', marker = 'o', linestyle = '--')
for i in range(len(x)):
    plt.text(x[i]-.2, y[i]+2, str(y[i]))
plt.text(0, 80, "A random graph", footsize = 12, color = 'blue')
```

```python
plt.plot(x,y)
plt.xlim(-5,15)
plt.ylim(20,80)
```

```python

plt.plot(x, y, color = 'red', marker = 'o', linestyle = '--', label = 'line 1')
plt.title("A random graph")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.legend()
plt.xticks(np.arange(0,11), rotation = 30)
plt.show()
```

```python
plt.figure(figsize=(7,4), dpi = 180, facecolor = '#efefef')
plt.plot(x, y, color = 'red', marker = 'o', linestyle = '--', label = 'line 1')
plt.title("A random graph")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.legend()
plt.xticks(np.arange(0,11), rotation = 30)
plt.show()
```

Graph Types
- Line Graph : show trends over time
- Bar Graph : compare values between categories(discrete data)
- Pie Chart : compare values between categories(percentage data)
- Histogram : show distribution of numerical dataa
- Box Plot : statistical distribution of numerical data
- Heat Map : show matrix of values using color
- Scatter Plot : show relationship between two variables
- Bubble Chart : show relationship between three variables

```python
#histogram
plt.hist('total_bill', data=tips, bins= range(0,60,10))
plt.show()
```

```python
h = plt.hist('total_bill', data=tips, bins= range(0,60,10),
             color = 'green', edgecolor= 'black',alpha = 5, hatch='//')
for i in range(len(h[0])):
    plt.text(x=h[1][i]+1, y=h[0][i]+2, s=str(int(h[0][i])))
plt.ylim(0,80)
plt.show()
```

```python
gender_count = tips.sex.value_counts()
gender_count
```

```python
plt.pie(gender_count.values,
        labels= gender_count.index,
        autopct= '%.1f%%')
plt.show()
```

```python
#donut
plt.pie(gender_count.values,
        labels= gender_count.index,
        autopct= '%.1f%%',
        wedgeprops= {'width':.3},
        pctdistance=.85,
        colors=['#ff9999','#66b3ff'])
plt.show()
```
