```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [0,1,2,3,4,5,6]
y = [45,32,56,76,87,54,32]
```

1. simple graph with automatic layout
2. manual layout

```python
plt.plot(x, y)
```

```python
# change graph line color, marker and style
plt.plot(x, y, color = 'red', marker = 'o', linestyle = '--')
```

```python
# add text for each point in the graph
plt.plot(x, y, color = 'red', marker = 'o', linestyle = '--')
for i in range(len(x)):
    plt.text(x[i]-.2, y[i]+2, str(y[i]))
plt.text(0, 80, "A random graph", footsize = 12, color = 'blue')
```

```python
plt.plot(x,y)
plt.xlim(-5,15)
plt.ylim(20,80)
```

```python

plt.plot(x, y, color = 'red', marker = 'o', linestyle = '--', label = 'line 1')
plt.title("A random graph")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.legend()
plt.xticks(np.arange(0,11), rotation = 30)
plt.show()
```

```python
plt.figure(figsize=(7,4), dpi = 180, facecolor = '#efefef')
plt.plot(x, y, color = 'red', marker = 'o', linestyle = '--', label = 'line 1')
plt.title("A random graph")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.legend()
plt.xticks(np.arange(0,11), rotation = 30)
plt.show()
```

manual graph setup and creation

```python
plt.subplots()
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("A random graph")
ax.set_xlabel("X Values")
ax.set_ylabel("Y Values")
ax.set_facecolor("#f9ff90")
ax.text(0,80, "A random graph", fontsize= 12, color= 'blue')
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=2)
ax1.plot(x, y)
ax1.set_title("A random graph")
ax1.set_xlabel("X Values")
ax1.set_ylabel("Y Values")
ax1.set_facecolor("#f9ff90")
ax1.text(0,80, "A random graph", fontsize= 12, color= 'blue')
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=2 , figsize =(12,4))
ax1.plot(x, y)
ax1.set_title("A random graph")
ax2.plot(x, z)
ax2.set_title("A random graph")
```

```python
fig, (ax1,ax2,ax3,ax4) = plt.subplots(nrows= 4, figsize=(4,12))
                                   
```

```python
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows = 2,
                                           ncols = 2, figsize=(8,8))
```

```python
fig, ax = plt.subplots()
#remove all the axis
ax.axis('off')
#remove the top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.scatter(y,z)
```

Graph Types
- Line Graph : show trends over time
- Bar Graph : compare values between categories(discrete data)
- Pie Chart : compare values between categories(percentage data)
- Histogram : show distribution of numerical dataa
- Box Plot : statistical distribution of numerical data
- Heat Map : show matrix of values using color
- Scatter Plot : show relationship between two variables
- Bubble Chart : show relationship between three variables
